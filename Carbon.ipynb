# Task-  Create a unified dataset that includes key project information (e.g., project type, location, credits issued, vintage).

```python
# Rename columns in df2 to align with df for merging
df2.rename(columns={'Project ID': 'ID', 'Project Site Location': 'Country/Area',
                    'Total Number of Offset Credits Registered': 'Estimated Annual Emission Reductions',
                    'Project Listed Date': 'Project Registration Date'}, inplace=True)
```

```python
print(df['ID'].dtype)
print(df2['ID'].dtype)
```

```python
# Convert ID column to string type for both dataframes
df['ID'] = df['ID'].astype(str)
df2['ID'] = df2['ID'].astype(str)
```

```python
unified_df = pd.merge(df, df2, on='ID', how='outer', suffixes=('_df', '_df2'))
```
